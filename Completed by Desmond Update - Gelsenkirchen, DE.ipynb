```python
#excercise 2: EmptyDict  

Inventory = {}
```

```python
#excercise 3: User Input 
Products = {"t-shirt": 15, "mug": 5, " hat": 8, "book": 3, "keychain": 12, "boots": 7, "slippers": 10, "glasses": 6, "toothbrush": 9, "toachbrush holder": 4, "mobile phone": 11}
Inventory = {}
for product, quantity in Products.items():
    Inventory[product] = quantity

print("Inventory:", Inventory)
```

```python
#excercise 4: Customer Orders

Purchases = []
customer_orders = set(Purchases)  
```

```python
#excercise 5: Customer Orders
Purchases = ["mug", "hat", "keychain",]
customer_orders = set(Purchases)  
```

```python
#excercise 6:Print Customer Orders
Purchases = ["mug", "hat", "keychain",]
customer_orders = set(Purchases)
print(customer_orders)  
```

```python
#excercise 7:Calculate order statistics
total_Purchases_ordered = len(customer_orders)
percentage_ordered = (total_Purchases_ordered / len(Products)) * 100
order_status = (total_Purchases_ordered, round(percentage_ordered, 2))
print(order_status) 
```

```python
#excercise 8:Calculate order statistics
print("\nPurchase Statistics:")
print(f"total_Purchases_ordered: {order_status[0]}")
print(f"Percentage of Purchases_ordered: {order_status[1]}%")
```

```python
# 9. Update inventory (subtract 1 for each ordered product)

for product in customer_orders:
    if product in Inventory:
        Inventory[product] -= 1
         
```

```python
# 10. Print updated inventory
for product in customer_orders:
    if product in Inventory:
        Inventory[product] -= 6
        print(f"Purchases ordered for {product}: {Inventory[product]}")  
```
